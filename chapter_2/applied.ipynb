### (a)

```python
auto.dtypes
```

```python
auto.origin.unique()
```

```python
auto.cylinders.unique()
```

```python
for i in auto.columns:
    if auto[i].nunique() < 20:
        if auto[i].dtypes == "float64" or auto[i].dtypes == "int64":
            print(i)
```

The variables mpg, displacement, weight, and acceleration are quantitative. The variables cylinders, year, origin, and name are qualitative.

### (b)

```python
quant_var = ["mpg", "displacement", "weight", "acceleration"]

for i in quant_var:
    min = np.min(auto[i])
    max = np.max(auto[i])
    print(f"{i} min: {min} | {i} max: {max}" )
```

### (c)

```python
for i in quant_var:
    mean = np.round(auto[i].mean(), 2)
    stdev = np.round(auto[i].std(), 2)
    
    print(f"{i} mean: {mean} | {i} standard deviation: {stdev}")
```

### (d)

```python
auto_subset = auto.drop(auto.index[10:86], axis = 0)

for i in quant_var:
    range = np.round(auto_subset[i].max() - auto_subset[i].min(), 2)
    mean = np.round(auto_subset[i].mean(), 2)
    stdev = np.round(auto_subset[i].std(), 2)
    print(f"{i} range: {range} | {i} mean: {mean} | {i} standard deviation: {stdev}")
```
